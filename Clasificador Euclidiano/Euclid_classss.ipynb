## Clasificador Euclidiano

La forma de encontrar la distancia es con base al famoso teorema de Pitagoras extendido a un espacio _n_-dimensional y que se conoce como _distancia euclidiana_; sin embargo, no es la unica forma de medir. 

Distancia _euclideana_. $r = $2.
<center>
$
d_{2}(\mathbf{x,y}) = [\sum_{i=1}^{n} |x_{i} - y_{i}|^2]^{1/2}
$
</center>

### Iris

```python
## contadores para cada clase
c1 = 0
c2 = 0
c3 = 0

c1_i = collections.Counter(entrena_HO_i[:,4])[0]
c2_i = collections.Counter(entrena_HO_i[:,4])[1]
c3_i = collections.Counter(entrena_HO_i[:,4])[2]

clas1_HO_i = np.zeros((c1_i,5))
clas2_HO_i = np.zeros((c2_i,5))
clas3_HO_i = np.zeros((c3_i,5))

for i in range(0, len(entrena_HO_i)):
    if entrena_HO_i[i,4] == 0 :
        clas1_HO_i[c1] = entrena_HO_i[i]
        c1 += 1
        
    elif entrena_HO_i[i,4] == 1 :
        clas2_HO_i[c2] = entrena_HO_i[i]
        c2 += 1
        
    elif entrena_HO_i[i,4] == 2 :
        clas3_HO_i[c3] = entrena_HO_i[i]
        c3 += 1
```

```python
pc1_i = np.mean(clas1_HO_i,0)
pc2_i = np.mean(clas2_HO_i,0)
pc3_i = np.mean(clas3_HO_i,0)
print(pc3_i)
```

```python
pred_i = np.zeros((len(prueba_HO_i),5))
buenas = 0
for i in range(0,len(prueba_HO_i)):
    ### clase1
    rasgo_1 = (prueba_HO_i[i,0] - pc1_i[0])**2
    rasgo_2 = (prueba_HO_i[i,1] - pc1_i[1])**2
    rasgo_3 = (prueba_HO_i[i,2] - pc1_i[2])**2
    rasgo_4 = (prueba_HO_i[i,3] - pc1_i[3])**2
    # 
    pred_i[i,0] = (rasgo_1+rasgo_2+rasgo_3+rasgo_4)**0.5
    ### clase 2
    rasgo_1 = (prueba_HO_i[i,0] - pc2_i[0])**2
    rasgo_2 = (prueba_HO_i[i,1] - pc2_i[1])**2
    rasgo_3 = (prueba_HO_i[i,2] - pc2_i[2])**2
    rasgo_4 = (prueba_HO_i[i,3] - pc2_i[3])**2
    # 
    pred_i[i,1] = (rasgo_1+rasgo_2+rasgo_3+rasgo_4)**0.5
    ### clase 3
    rasgo_1 = (prueba_HO_i[i,0] - pc3_i[0])**2
    rasgo_2 = (prueba_HO_i[i,1] - pc3_i[1])**2
    rasgo_3 = (prueba_HO_i[i,2] - pc3_i[2])**2
    rasgo_4 = (prueba_HO_i[i,3] - pc3_i[3])**2
    # 
    pred_i[i,2] = (rasgo_1+rasgo_2+rasgo_3+rasgo_4)**0.5
    
    menor = pred_i[i,0]
    for j in range(3):
        if pred_i[i,j] < menor:
            menor = pred_i[i,j]
        
    if (menor == pred_i[i,0]):
        pred_i[i,3] = 0
    
    elif(menor == pred_i[i,1]):
        pred_i[i,3] = 1
    
    elif(menor == pred_i[i,2]):
        pred_i[i,3] = 2
    
    if(prueba_HO_i[i,4] == pred_i[i,3]):
        pred_i[i,4] = 1
        buenas += 1
    else:
        pred_i[i,4] = 0
    
print("El resultado de calculo de medias es \n {}".format(pred_i))
print("Se predijeron correctamente {} patrones y se tuvo un porcentaje de: {:.2f} %".format(buenas,(buenas/len(prueba_HO_i))*100))
```

# K-Fold Cross Validation

Se crean 10 folds que contienen 9 subconjuntos o folds como entrenamiento

```python
lon_K_i = int(len(iri_p)/10)
lon_K_h = int(len(haber_p)/10)
lon_K_d = int(len(diab_p)/10)
```

## Iris K-Fold

```python
pruebaK_i = np.zeros((lon_K_i,4))
entrenaK_i = np.zeros((lon_K_i*9,4))
iri_p = iri_p.astype(float)
media = np.zeros((10))
buenas_ki = np.zeros((10))

for i in range(0, 10):
    k = i *lon_K_i
    pruebaK_i = iri_p[k:k+lon_K_i,:]
    entrenaK_i = np.concatenate((iri_p[:k], iri_p[k+lon_K_i:]), axis =0)
    
    c1_i = collections.Counter(entrenaK_i[:,4])[0]
    c2_i = collections.Counter(entrenaK_i[:,4])[1]
    c3_i = collections.Counter(entrenaK_i[:,4])[2]
    clas1_K_i = np.zeros((c1_i,5))
    clas2_K_i = np.zeros((c2_i,5))
    clas3_K_i = np.zeros((c3_i,5))
    c1 = 0
    c2 = 0
    c3 = 0
    
    for j in range(0, len(entrenaK_i)):
        if entrenaK_i[j,4] == 0 :
            clas1_K_i[c1] = entrenaK_i[j]
            c1 += 1
        
        elif entrenaK_i[j,4] == 1 :
            clas2_K_i[c2] = entrenaK_i[j]
            c2 += 1
        
        elif entrenaK_i[j,4] == 2 :
            clas3_K_i[c3] = entrenaK_i[j]
            c3 += 1
        
    ## Calculo de las medias por clase
    pc1_i = np.mean(clas1_K_i,0)
    pc2_i = np.mean(clas2_K_i,0)
    pc3_i = np.mean(clas3_K_i,0)
    
    pred_i = np.zeros((len(pruebaK_i),5))
    buenas = 0
    
    ## Evaluacion
    for l in range(0,len(pruebaK_i)):
        ### clase1
        rasgo_1 = (pruebaK_i[l,0] - pc1_i[0])**2
        rasgo_2 = (pruebaK_i[l,1] - pc1_i[1])**2
        rasgo_3 = (pruebaK_i[l,2] - pc1_i[2])**2
        rasgo_4 = (pruebaK_i[l,3] - pc1_i[3])**2
        # 
        pred_i[l,0] = (rasgo_1+rasgo_2+rasgo_3+rasgo_4)**0.5
        ### clase 2
        rasgo_1 = (pruebaK_i[l,0] - pc2_i[0])**2
        rasgo_2 = (pruebaK_i[l,1] - pc2_i[1])**2
        rasgo_3 = (pruebaK_i[l,2] - pc2_i[2])**2
        rasgo_4 = (pruebaK_i[l,3] - pc2_i[3])**2
        # 
        pred_i[l,1] = (rasgo_1+rasgo_2+rasgo_3+rasgo_4)**0.5
        ### clase 3
        rasgo_1 = (pruebaK_i[l,0] - pc3_i[0])**2
        rasgo_2 = (pruebaK_i[l,1] - pc3_i[1])**2
        rasgo_3 = (pruebaK_i[l,2] - pc3_i[2])**2
        rasgo_4 = (pruebaK_i[l,3] - pc3_i[3])**2
        # 
        pred_i[l,2] = (rasgo_1+rasgo_2+rasgo_3+rasgo_4)**0.5
    
        menor = pred_i[i,0]
        for j in range(3):
            if pred_i[l,j] < menor:
                menor = pred_i[l,j]
        
        if (menor == pred_i[l,0]):
            pred_i[l,3] = 0
    
        elif(menor == pred_i[l,1]):
            pred_i[l,3] = 1
    
        elif(menor == pred_i[l,2]):
            pred_i[l,3] = 2
    
        if(pruebaK_i[l,4] == pred_i[l,3]):
            pred_i[l,4] = 1
            buenas += 1
        else:
            pred_i[l,4] = 0

    media[i] = buenas/(lon_K_i) * 100
    buenas_ki[i] = buenas

print("El resultado de calculo de medias es \n {}".format(media))
print("El numero de predicciones correctas por fold fue: {} ".format(buenas_ki))
print("y se tuvo un porcentaje promedio  de: {:.2f} %".format(np.mean(media)))
print("Cada fold contiene: {} patrones".format(lon_K_i))
```
