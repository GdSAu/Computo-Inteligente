```python
import numpy as np
import random
from sklearn.model_selection import train_test_split, KFold, LeaveOneOut
```

## Creacion del array

```python
# Crear el array de 100 filas con 2 columnas con 0s
q = np.zeros((100,2))

#Se recorrera el arreglo con la finalidad de "rellenarlo"
#Se emplea la funcion random.randint() para asignar un numero entero que este dentro de un rango
#El rango para este caso sera 0-9 y se asignaran
#45, 35 y 25 datos a las 3 clases requeridas
for i in range(0,100):
    if i < 40: #Clase1
        q[i][0] = random.randint(0,9)
        q[i][1] = 1
    elif  i >= 40 and i < 65 :#Clase2
        q[i][0] = random.randint(0,9)
        q[i][1] = 2
    elif i >= 65:#Clase3
        q[i][0] = random.randint(0,9)
        q[i][1] = 3
q 
```

```python
#Se realiza una permutacion aleatoria de los datos generados
q = np.random.permutation(q)
q
```

```python
## Verificar si los datos no se han modificado, se debe tener 40-1's, 35-2's y 25-3's
n = 0
b = 0
c = 0 

for i in range (0,100):
    
    if q[i][1] == 1:
        n = n+1
    
    elif q[i][1] == 2:
        b = b+1
    elif q[i][1] == 3:
        c = c+1
print(n)
print(b)
print(c)
```

## Hold-Out

```python
# Se realiza la division del set por el metodo HOLD OUT
train,test = train_test_split(q,train_size =0.6, test_size=0.4)
# Verificamos que cada uno de los conjuntos contenga el porcentaje requerido
print(len(train),len(test))
```

```python
print('Los datos de entrenamiento son:\n')
print(train)
print("Los datos de validacion son:\n ")
print(test)
 
```

## K-Fold

```python
#Implementacion del metodo K-Fold
#se inicializa el numero de folds en 10
kf = KFold(n_splits= 10)
kf.get_n_splits(q)
print(kf)
```

```python
z = 0
for train_index, test_index in kf.split(q):

    X_train, X_test = q[train_index], q[test_index]
    print("\nFold: %d"%(z))
    print("TRAIN:\n")
    for i in range (0,len(train_index)):
        print(X_train[i])
    print("TEST:\n")
    for i in range (0,len(test_index)):
        print(X_test[i])
    z = z+1
```

## Leave-One-Out

```python
# Implementacion del metodo Leave-One-Out
loo = LeaveOneOut()
loo.split(q)
```

```python
#Inicia
z = 0
for train_ix, test_ix in loo.split(q):
	# split data
    #print("TRAIN:", train_ix, "TEST:", test_ix)
    print("Set:%d"%(z))
    X_train, X_test = q[train_ix, :], q[test_ix, :]
    print("Train:\n")
    for i in range (len(X_train)):
        print(X_train[i])
    print("Test:")
    print(X_test)
    z = z+1
```
